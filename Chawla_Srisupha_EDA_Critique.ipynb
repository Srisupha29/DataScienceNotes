Exploratory Data Analysis (EDA) plays a critical role in understanding datasets before performing predictive modeling or building data-driven applications. The original exploratory analysis of the US Tech and EV stock dataset demonstrates several useful initial steps, but it also contains notable limitations that reduce the depth of insights produced.


**Strengths:** One of the main strengths of the original EDA is its basic structural exploration of the dataset. The analysis successfully identifies the dataset dimensions and inspects the dataset structure using common exploratory commands such as .head() and .info(). These steps help verify that the dataset contains the expected variables including Date, Ticker, Open, Close, High, Low, and Volume. Additionally, the dataset structure was reorganized and cleaned to ensure that columns were correctly formatted, allowing the data to be used for time-series analysis. This demonstrates an understanding of the importance of data preparation before conducting further analysis.
Another positive aspect of the original EDA is the generation of descriptive statistics for numerical variables. Using functions such as .describe(), the analysis provides summary statistics including mean, minimum, maximum, and standard deviation for stock prices and trading volume. These statistics offer a general overview of the dataset and help identify potential anomalies or unusual values within the price ranges.


Despite these strengths, the original EDA lacks several essential components that are necessary for a comprehensive exploratory analysis. One major gap is the absence of a clear research objective. While the dataset itself suggests that the analysis relates to stock market behavior, the original analysis does not explicitly state a research question such as examining price volatility, comparing stock performance across companies, or identifying factors that influence trading activity. Without a defined objective, the analysis lacks a clear direction.


**Weaknesses:** Another significant weakness is the limited assessment of data quality. Although the dataset appears to be relatively clean, the original EDA does not systematically check for duplicate rows, analyze patterns of missing values, or perform formal outlier detection. These steps are essential because unaddressed data quality issues can distort statistical summaries and negatively affect downstream machine learning models.


The original analysis also lacks sufficient visual exploration of relationships between variables. For example, correlations between stock prices, trading volume, and other numerical variables are not evaluated. Similarly, comparisons between companies using visual tools such as boxplots or grouped analyses are missing. Without these analyses, important patterns such as price volatility differences between companies remain unexplored.
The lack of feature engineering is another limitation. The dataset contains raw price values, but additional informative features such as daily returns, price range, or temporal components derived from the date variable were not created. These derived features can significantly improve the analytical power of the dataset and provide more meaningful insights into stock performance.


**Impact:** These gaps reduce the depth of the analysis and limit the ability to draw strong conclusions about stock behavior. If these missing components remain unaddressed, any predictive models built using this dataset may perform poorly due to insufficient understanding of the underlying data structure.

**Improvements:**  To improve the analysis, several enhancements can be implemented. First, additional data quality checks should be conducted, including duplicate detection, missing value visualization, and outlier identification. Second, more comprehensive visualizations should be introduced to explore distributions and relationships between variables. Third, new features such as daily returns, price volatility, and date-based attributes should be engineered to enrich the dataset. Finally, statistical analyses such as correlation analysis and hypothesis testing should be performed to better understand relationships between variables.
By incorporating these improvements, the EDA can provide a more thorough understanding of the dataset and create a stronger foundation for subsequent predictive modeling and data visualization tasks.

### 5. Missing Component: Feature Engineering
New features were not created in the original EDA.


```python
df_final['Daily_Return'] = (df_final['Close'] - df_final['Open']) / df_final['Open']
df_final['Price_Range'] = df_final['High'] - df_final['Low']
df_final['Price_Ticker'] = pd.to_datetime(df_final['Price_Ticker'])
df_final['Year'] = df_final['Price_Ticker'].dt.year
df_final['Month'] = df_final['Price_Ticker'].dt.month
```

 ### 6. Missing Component: Memory Usage Assessment

### Interpretation
Feature engineering provides additional variables that help analyze market volatility and price movements more effectively.

### Statistical Analysis 1: Correlation Matrix

```python
corr_matrix = df_final[['Open','Close','High','Low','Volume']].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix of Stock Variables")
plt.show()
```

### Interpretation
Correlation analysis reveals relationships between stock price variables and trading volume.

**Correlation Matrix of Stock Variables**

The correlation heatmap reveals extremely strong positive correlations among the price-related variables (Open, Close, High, and Low), with correlation values close to 1.  

This is expected because these variables all describe the price of the same stock within the same trading day.  

In contrast, the Volume variable shows a moderate negative correlation with price variables (around -0.44 to -0.46). This suggests that higher trading volumes tend to occur when prices fluctuate more, although the relationship is not extremely strong.

### Statistical Analysis 2: Hypothesis Testing
Statistical Analysis 2: Hypothesis Testing 

```python
from scipy.stats import ttest_ind

tesla = df_final[df_final['Ticker']=='TSLA']['Daily_Return']
apple = df_final[df_final['Ticker']=='AAPL']['Daily_Return']

ttest_ind(tesla, apple)
```

### Interpretation
This test evaluates whether the average daily returns of Tesla and Apple differ significantly.
