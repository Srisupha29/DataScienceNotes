````python
# Create a 3-d array with data

n = [ [[1,2,3],
       [4,5,6]],
     
     [[23, 24, 25],
      [26, 27, 28]],

     [[66,65,64],
      [63,62,61]]
    ]
       

m = [
    
    [[1,2,3],
     [4,5,6]],
    
      [[10,20,30],
      [40,50,60]],
    
     [[100,200,300],
      [400,500,600]]
]

assert m[0][0][0] == 1
assert m[1][1][0] == 40
assert m[0][0][1] == 2
assert m[2][1][2] == 600
assert n[2][0][0] == 66
assert n[0][1][2] == 6
assert m[-2:-1][0] == [[10,20,30], [40,50,60]]
assert m[-1][0:] == [[100, 200, 300], [400, 500, 600]]

# m[-1:-2][0] # IndexError: list index out of range

assert m[-2:-1][0] ==  [[10,20,30],[40,50,60]]
# m[-2:-1] gets [ [[ 10,20,30], [40,50,60]] ] --[0]--> [[10,20,30],[40,50,60]]

assert m[-1][0] == [100,200,300]

assert m[-1:][0] == [[100, 200, 300], [400, 500, 600]]   
# m[-1:] gets [ [[100,200,300],[400,500,600]] ] --[0]--> [[100, 200, 300], [400, 500, 600]]   

assert m[-1:][:0] == [] 
assert m[-1:][0:] == [[[100, 200, 300], [400, 500, 600]]]
# m[-1:] gets [ [[100,200,300],[400,500,600]] ] -->[0:] --> gets the exact same thing 
````

### Situations where Such Methods are Applicable

````python
# create a dataframe to be used next
addr1_df = pd.DataFrame({
    'Number' : ['114', '153/1', '102', '945'],
    'Building' : ['A', 'BeachFront', 'Main', 'C'],
    'Street' : ['Sai 1', 'Rama IX', 'Barom' , 'Yaek 9'],
    "City":["Salaya", "Payathai", "Salaya", "Aroonamarin"],
    "District":["Taweewattana", "Ratchtevee", "Taweewattana", "Bangkok Noi"],
    "Province":["Bangkok", "Bangkok", "Bangkok", "Bangkok"],
    "ZIP":[73710, 10400, 73710, 10700]})

addr1_df
                         
````

````python
# One to one merge
visited_subset = visited.loc[[0, 2, 6]]
visited_subset
````

````python
# both the dataframes above have similar content but the content is under different names
# we can merge based on that

# if just do this, merge wont work since we dont have any common column to perform on
one2one_merge = site.merge(visited_subset)
one2one_merge

# Merge error
````

````python
# now, name and site columns have the the same data but are set with different column names
# we use these to merge the dataframes together
one2one_merge = site.merge(visited_subset, left_on = 'name', right_on = 'site')
one2one_merge
````

````python
# we can switch as well
one2one_merge = visited_subset.merge(site, left_on = 'site', right_on = 'name')
one2one_merge
````

````python
# however, we can use join to get a similar result 
# but with Nan values since site does not have index 6
one2one_join = visited_subset.join(site)
one2one_join
````

````python
# since visited_subset only covers [[0,2,6]] -> row 1 shows nan values for visited_subset
one2one_join = site.join(visited_subset) 
one2one_join
````

````python
# same data but in site df, its called "name"
# in visited df, its called "site"
Many2One_merge = visited.merge(site, left_on = 'site', right_on = 'name')
Many2One_merge                         
````

````python
# Same concept as given above, we just swap the order
One2Many_merge = site.merge(visited, left_on = 'name', right_on = 'site')
One2Many_merge                
````

````python
# Manyto Many Merge
enrollment_df = pd.DataFrame({
    'ID' : [1001, 1002],
    'Course' : ["ICCS161", "ICCS161"]
})
enrollment_df
````

````python
exam_df = pd.DataFrame({
    "Course": ["ICCS161", "ICCS161"],
    "Exam": ["Midterm", "Final"]
})
exam_df
````

````python
Many2Many_merge = enrollment_df.merge(exam_df)
Many2Many_merge                                     
````

````python
enrollment_df_2 = pd.DataFrame({
    "ID": [1001, 1002, 1003],
    "Course": ["ICCS161", "ICCS161", "ICCS331"]
})
enrollment_df_2
````

````python
# since ICCS331 is not a match in exam_df, it doesnt show up at all
Many2Many_merge_2 = enrollment_df_2.merge(exam_df)
Many2Many_merge_2
````

````python
# change the name of the index and merge on that basis
enrollment_df_2.set_index('Course', inplace = True)
enrollment_df_2
````

````python
exam_df.set_index('Course', inplace = True)
exam_df
````

````python
# since we are using "outer" , even 1003 shows up but just with a Nan value
joiney = enrollment_df_2.join(exam_df, how = 'outer')
joiney                     
````

````python
# with inner, only the common values show up
joiney2 = enrollment_df_2.join(exam_df, how = 'inner')
joiney2   
````

### Conclusion - Similarities and Differences between Concat, Merge, and Join

````python
df1 = pd.read_csv("/Users/srisuphachawla/Documents/DataScienceDownloads/concat_1.csv")
df2 = pd.read_csv("/Users/srisuphachawla/Documents/DataScienceDownloads/concat_2.csv")
````

````python
df3 = df2.copy()

df3.iloc[2] = df1.iloc[1]
df3.iloc[3] = df1.iloc[2]

df2.columns = ['D', 'E', 'F', 'G']
df3.columns = ['A', 'D', 'G', 'H']

df2.index=[0, 2, 3, 5]
df3.index=[2, 3, 6, 7]
````

````python
df1
````

````python
df2
````

````python
df3
````

````python
pd.concat([df1, df2])
````

````python
pd.concat([df1, df2], axis =1)
````

````python
pd.concat([df1, df2], join = 'inner')
````

````python
pd.concat([df1, df2], axis = 1, join = 'inner')
````

````python
df1.join(df2, lsuffix="_left", rsuffix="_right", how ='outer')
````

````python
pd.merge(df1, df2, left_index=True, right_index=True)
````

````python
# based on left - df1 indexes only
df1.join(df2, lsuffix="_left", rsuffix="_right", how='left')
````

## Missing Data

### Types of Missing Data
- **Missing completely at random (MCAR)**
  - Unrelated to both<br>a) the specific value that should have been obtained<br>**and**<br>b) the set of observed responses
    - No systematic differences exist between participants with missing data and those with complete data
  - Samples are lost accidentally or due to technical problems
  - *Pro*: Power could be lost but estimated parameters are unbiased.
  - *Con*: Don't be too fast to assume this case.
- **Missing at random (MAR)**
  - Systematically related to the observed but not the unobserved data
    - The probability that a value is missing depends on observed values of other variables
      - not on the other missing values of the target variable.
  - Example: the income level of a client can be estimated from other variables like the clients profession, experience and qualification
- **Missing not at random (MNAR)**
  - The data doesn't fit the creiteria of MCAR and MAR
  - Example: Citizens of a country opt not participate in a survey
  

3. Handling Missing Data

- Plan a study really carefully
  - Easier said than done
- Methods to handle missing data
  - Do nothing
  - Use only complete observaions
  - Use single imputation
  - Use multiple imputation

# Handling Missing Data  MCAR, MAR, MNAR

Missing data is a common problem in real-world datasets.  
Understanding *why* data is missing is very important because it affects:

- how we analyze data  
- whether we can safely drop rows  
- whether imputation is valid  

There are three main statistical types of missing data:

1. MCAR  Missing Completely At Random  
2. MAR  Missing At Random  
3. MNAR  Missing Not At Random  

---

## 1. MCAR  Missing Completely At Random

### Definition

Data is MCAR if the probability that a value is missing is **independent of all variables**  
(both observed and unobserved).

### Interpretation

- Missingness happens purely by chance  
- No pattern in missing values  
- Like a random accident

### Example 1  Technical Failure

A survey dataset loses some responses because:

- internet connection failed  
- a computer crashed  

These missing values are unrelated to age, gender, income, etc.

 This is MCAR.

### Example 2  Lab Accident

A researcher accidentally drops some test tubes and loses 5 samples.

- The lost samples are random  
- Not related to any measurement  

 MCAR

### Important Property

If data is MCAR:

- Dropping missing rows does **not** introduce bias  
- Simple methods like listwise deletion are acceptable  

---

## 2. MAR  Missing At Random

### Definition

Missingness depends on **observed variables**,  
but not on the actual missing value itself.

\[
P(\text{missing}) \text{ depends only on known variables}
\]

### Key Idea

We can explain the missingness using other columns in the dataset.

---

### Example 1  Income Survey

Columns collected:

- Age  
- Gender  
- Income  

Observation:

- Younger people are more likely to skip the income question

So:

- Missing income depends on AGE (which we know)

 This is MAR

---

### Example 2  Medical Study

Older patients are less likely to return for follow-up blood pressure checks.

- Age is observed  
- Missing BP depends on age  

 MAR

---

### Example 3  Education Survey

Men are less likely to answer mental health questions.

- Missingness depends on gender (observed)

 MAR

---

### Why MAR Is Important

Under MAR:

- Statistical imputation methods are usually valid  
- Methods like regression imputation or multiple imputation work well  

---

## 3. MNAR  Missing Not At Random

### Definition

Missingness depends on the **actual value that is missing**,  
even after accounting for observed variables.

\[
P(\text{missing}) \text{ depends on the missing value itself}
\]

---

### Example 1  Salary Data

People with very high salaries refuse to report income.

- The probability of missing income increases with income

 MNAR

---

### Example 2  Mental Health Survey

People with severe depression avoid answering questions about mental health.

- Missingness depends on severity of depression

 MNAR

---

### Example 3  Weight Reporting

Overweight individuals avoid reporting their weight.

- Missing weight is directly related to true weight

 MNAR

---

### Why MNAR Is Difficult

- Dropping rows causes bias  
- Simple imputation is usually incorrect  
- Often requires domain knowledge or special models  

---

# Summary Table

| Type | Depends On | Example |
|-----|-----------|---------|
| MCAR | Nothing | Random computer failure |
| MAR | Observed variables | Young people skip income |
| MNAR | The missing value itself | Rich people hide salary |

---

# Practical Consequences

| Type | Can Drop Missing Rows? | Imputation Safe? |
|-----|------------------------|-----------------|
| MCAR | Yes | Yes |
| MAR | No | Yes (with good methods) |
| MNAR | No | Very risky |

---

# Memory Trick

- **MCAR**  Pure bad luck
- **MAR**  Missing explained by other columns
- **MNAR**  Missing because of the value itself

---
