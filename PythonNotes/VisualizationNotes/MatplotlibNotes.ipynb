```python
import matplotlib.pyplot as plt
import numpy as np 
from matplotlib import style
```

# Scatter Plot

```python
# scatter plots: will result in a new chart everytime we run it given that we are using random numbers

x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

print(x_data) # we can even check our x or y values
plt.scatter(x_data, y_data, c = "#0f0") # we can specify colours in different ways.
plt.scatter(x_data, y_data, c = "green", marker = "*") # marker can be used to change shapes
plt.scatter(x_data, y_data, c = "blue", s = 66) # s = n can be used for size
plt.show()
```

```python
x_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100

plt.scatter(x_data, y_data, c="purple", s=50, marker ="*")
# we can see, that with 1000 numbers, the data is overlapping 

plt.scatter(x_data, y_data, c = "#0f0", alpha= 0.3)
# alpha allows some transparency 

plt.show()
```

# Histogram

```python
ages = np.random.normal(20, 1.5, 1000)
# generates 1000 random numbers following a Gaussian distribution with a mean of 20 and std of 1.5
plt.hist(ages)
# we can also specify the bins (the block-like structures on the graph) 

plt.hist(ages,
         bins = 5, color = 'purple') # this will give us 5 bins

# However, if we want to specify where and till which point we want the bins to be
 
plt.show()
```

```python
# However, if we want to specify where and till which point we want the bins to be
plt.hist(ages, 
         bins = [ages.min(), 18, 21, ages.max()], color = 'teal')
# here, from the min.age to 18 --> bin 1
#       from 18 to 21 -----------> bin 2
#       from 21 to max.age ------> bin 3

# hence, we get to see 3 bins

plt.show()
```

```python
# shows the data of 1000 people 
# Each bar's height = the number of data points up to and including that bin.
# This makes it easy to see how the data accumulates over the range.
# The last bar will have the total number of observations.

plt.hist(ages, 
         bins = 20,
         cumulative = True) 

plt.show()
```

# Box Plots

```python
height = np.random.normal(172, 8, 300)

plt.boxplot(height)
```

How to read a BoxPlot?
- the dots outside the box represent the outliers
- The horizontal line inside the box is around 170 cm, the orange line, which is our median
- A box plot is separated into 3 quartiles.
       - Q1(25%): Bottom of the box, 167 cm, 25% of the people are shorter than 167 cm
       - Q2(50%): our median, 50% of people are shorter than 170 cm
       - Q3(75%): Top of the box, 175 cm, 75% people are shorter than 175 cm

- The straight lines are known as "Whiskers", shows regularity:
       - Lower whisker comes down to 150 cm: shortest "regular" height in the dataset
       - Upper whisker goes up till 193 cm: tallest "regular" height in the dataset
 
- Anything beyond whiskers are considered outliers, ie, the small dots outside
   - Below: several values are below 150 cm, two in particular which are below 145 cm
   - Above: two values slightly above 193 cm

- The median is slightly below the center of the data:
      - mild right skew (more variations in the taller heights)

```python
first = np.linspace(0, 10, 25)
# creates an array of 25 numbers evenly spaced from 0 to 10
sec = np.linspace(10, 200, 25)
# creates an array of 25 numbers evenly spaced from 10 to 200
third = np.linspace(200, 210, 25)
# creates an array of 25 numbers evenly spaced from 200 to 210
fourth = np.linspace(210, 230, 25)
# creates an array of 25 numbers evenly spaced from 210 to 230

data = np.concatenate((first, sec, third, fourth))
plt.boxplot(data)
plt.show()


# we can see here that median itself has a huge space, 
# thats because our second array contains a wide range, where the rest of them have a 
# smaller range. 

# This does not change the fact that each array is made up of the same amount of numbers
```

# Plot Customization

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
income = [55,56,62,61,72,72,73,75]

income_tickets = list(range(50, 81, 2))
# range(50, 81) includes all the numbers within our income
# the 2 is for spacing

plt.plot(years, income)

# we want to have a title and basic explanation for easy understanding 
plt.title('Income of John Doe (in USD)', fontsize = 2)
plt.xlabel("Year") #specifies the x axis
plt.ylabel("Yearly Income (USD)") #specifies the y axis
plt.yticks(income_tickets, [f"{x}k USD" for x in income_tickets])


plt.show()
```

# Legends and Multiple Plots

```python
# We can also show multiple lines in one graph 

# you can also see that when you only provide one side, the other side will automatically 
# generate numerical values.

style.use("dark_background") # importing style, customize the background

a = [32,56,21,67,43,69]
b = [85,21,34,22,33,78]
c = [100,12,32,90,2,88]

plt.plot(a, label = "AA" )
plt.plot(b, label = "BB" )
plt.plot(c, label = 'CC' )

plt.legend() # writing the label itself wont do anything but .legend() will add 
             # the detailing of each line

plt.legend(loc="upper right") # we can also change the position 
# this method works with other charts as well. 

plt.show()
```

# Subplots

```python
x = np.arange(100)
fig , axs = plt.subplots(2,2)
# for a 2x2 grid that has 4 subplots 

# [row, column] representation

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sine Wave")


axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("CoSine Wave")


axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title("Random Wave")


axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("Log Wave")
axs[1,1].set_xlabel("Test") 
# we can see that the corner most, ie, row 1 col 1, has Test as the name of its x axis

fig.suptitle("Four Plots") # super title: name 

plt.tight_layout() # prevents overlapping of names 

plt.savefig("fourplots.png", dpi = 300,
           transparent = True, bbox_inches = 'tight') 
# .savefig() allows us save this, making it show up in our system
# dpi is for a higher quality image
# transparency = True, 
```

# 3D plotting

```python
ax = plt.axes(projection = "3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)

ax.set_title("3D plot")

plt.show()

# we can also set x, y, or x names through set_xlabel 
```

```python
ax = plt.axes(projection = "3d")

x = np.arange(0, 50, 0.1)
y = np.arange(0, 50, 0.1)
z = np.cos(x+ y )

ax.scatter(x, y, z)

ax.set_title("3D plot")

plt.show()
```

```python
# How can we plot actual surfaces?
ax = plt.axes(projection = "3d")

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y )

Z = np.sin(X) * np.cos(Y)

ax.plot_surface( X, Y, Z, cmap = "Spectral") # cmap is for color

#      ax.view_init(azim=1, elev = 90) # sets the upper view as default
ax.set_title("3D plot")

plt.show()
```

# Animations in Plots

```python
import random 

heads_tails =[0,0]

for _ in range(100000):
    heads_tails[random.randint(0,1)] += 1
    plt.bar(['Heads', 'Tails'], heads_tails, color = ['red', 'blue'])
    plt.pause(0.001) # used for animating functions
  

plt.show()
```

plt.bar()  makes vertical bars (X-axis = categories, Y-axis = values).
plt.barh()  makes horizontal bars (Y-axis = categories, X-axis = values).
