```python
import numpy as np
import pandas as pd
```

```python
# defining the given items
N = 500000
a = 2
ratios = []

for i in range(10):
    # generating standard normal samples 
    X = np.random.randn(N)
    W = X**2

    # computing ration
    ratio = np.sum(W <= a) / N
    ratios.append(ratio)

min_ratio = min(ratios) 
max_ratio = max(ratios)
print("Ratios: ", ratios)
print("Min ratio: ", min_ratio)
print("Max_ratio: ", max_ratio)
print("Check if min_ratio < 0.8427 < max_ratio: ", min_ratio < 0.8427 < max_ratio)
```
